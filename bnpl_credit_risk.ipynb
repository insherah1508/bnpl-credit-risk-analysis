```python
import pandas as pd

data = pd.read_csv("creditcard.csv")
data.head()
```

```python
data["Class"].value_counts()
```

```python
df = data.copy()
df = df.rename(columns={"Class": "bad_outcome", "Amount": "purchase_amount"})
df[["purchase_amount", "bad_outcome"]].head()
```

```python
df["approved"] = df["purchase_amount"] <= 200
df[["purchase_amount", "approved", "bad_outcome"]].head()
```

```python
df["approved"].value_counts(normalize=True)
```

```python
approved_df = df[df["approved"] == True]
approved_df["bad_outcome"].mean()
```

```python
df["approved_500"] = df["purchase_amount"] <= 500

approval_rate_500 = df["approved_500"].mean()

approved_500_df = df[df["approved_500"] == True]
bad_rate_500 = approved_500_df["bad_outcome"].mean()

approval_rate_500, bad_rate_500
```

```python
df["approved_all"] = True

approval_rate_all = df["approved_all"].mean()

bad_rate_all = df["bad_outcome"].mean()

approval_rate_all, bad_rate_all
```

## Business Recommendation

Increasing BNPL purchasing power from $200 to $500 significantly improves approval rates
while maintaining a stable bad outcome rate. Approving all purchases introduces additional
risk without proportional benefit.

Recommendation: Implement capped purchasing power to balance growth and credit risk.
